```python
filelist= os.listdir('voices//ourvoices//Esraa')
df_our3 = pd.DataFrame(filelist)
df_our3['label']='Esraa'
df_our3 = df_our3.rename(columns={0:'file'})
df_our3
```

```python
filelist = os.listdir('voices//ourvoices//MaryamMegahed')
df_our2 = pd.DataFrame(filelist)
df_our2['label']='Mariam'
df_our2 = df_our2.rename(columns={0:'file'})
```

```python
df_our
```

```python
filelist=os.listdir('voices//others//Diaa')
filelist+=os.listdir('voices//others//Saeed')
filelist+=os.listdir('voices//others//DinaHussam')
filelist+=os.listdir('voices//others//Romaisaa')
filelist+=os.listdir('voices//others//Shaban')
filelist+=os.listdir('voices//others//Kamel')
filelist+=os.listdir('voices//others//Sama')



#read them into pandas
df_others = pd.DataFrame(filelist)
df_others['label']='0'
df_others = df_others.rename(columns={0:'file'})
df_others
```

```python
df = pd.concat([df_our, df_others,df_our1,df_our2,df_our3], ignore_index=True)
df.head()
#df.drop(0,inplace=True)

```

```python
# for i in range(0,len(df)):
#     try:
#         df[i:i+1].apply(extract_features, axis=1)
#     except:
#         df.drop(df.loc[i].name,inplace=True)         

```

```python
features_label = df.apply(extract_features, axis=1)
features=feat(features_label)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features,df['label'], test_size=0.3, random_state=1)
```

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=5)
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
